```python
import numpy as np
import pandas as pd
import cv2
import os
from IPython.display import display, Image

OUTDIR = '/data/work/liver_hcc_spatial/01_gem2png'
os.makedirs(OUTDIR, exist_ok=True)

SAMPLES = {
    'Y10393D8': 'W202603240063157',
    'Y40320PC': 'W202603240063121',
    'Y40320ND': 'W202603240063078',
    'Y10052H8': 'W202603240063031',
    'Y10052E3': 'W202603240062997',
    'Y10052F1': 'W202603240062964',
    'Y10052H4': 'W202603240062927',
    'Y01647G5': 'W202603240062892',
    'Y01647G8': 'W202603240062792',
}

for sample, wfid in SAMPLES.items():
    gem_path = f'/data/input/Files/ResultData/Workflow/{wfid}/result/{sample}.raw.gem'
    out_path = f'{OUTDIR}/{sample}.mRNA.png'
    
    print(f'Processing {sample} ...', end=' ')
    data = pd.read_csv(gem_path, sep='\t', comment='#')
    data['x'] -= data['x'].min()
    data['y'] -= data['y'].min()
    data = data.groupby(['x', 'y']).sum().reset_index()
    w = data['x'].max() + 1
    h = data['y'].max() + 1
    mtx = np.zeros((w, h))
    mtx[data['x'], data['y']] = data['MIDCount']
    mtx *= 250
    mtx = np.clip(mtx, 0, 255)
    cv2.imwrite(out_path, mtx)
    print(f'done → {out_path}')

print('\nAll done!')
```
