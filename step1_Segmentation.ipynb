```python
import sys
sys.path.append('/data/work/liver_hcc_spatial/xxx')
from Tools.Segmentation import Segmentation
import matplotlib.pyplot as plt
from Tools.Spatial import *
import numpy as np
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
samples = pd.read_csv('/data/work/liver_hcc_spatial/segmentation_samples.csv')
outputdir = '/data/work/liver_hcc_spatial/03_segmentation/'
os.makedirs(outputdir, exist_ok=True)

for _, row in samples.iterrows():
    print(f'\n========== 开始处理: {row["id"]} ==========')
    
    seg = Segmentation()
    seg.load(img_path=row['ssDNA_path'], mRNA_path=row['gem_path'])
    
    seg.pre_process(threshold='auto', verbose=False)
    
    seg.watershed(
        block_size=int(row['block_size']),
        offset=float(row['offset']),
        min_distance=int(row['min_dist']),
        expand_distance=int(row['expand_distance']),
        verbose=False
    )
    
    seg.save_scGEM(save_path=outputdir, name=row['id'], verbose=True)
    
    # 保存分割可视化图
    fig, axes = plt.subplots(1, 2, figsize=(20, 10))
    axes[0].imshow(seg.raw_img, cmap='gray')
    axes[0].set_title(f'{row["id"]} - Raw')
    axes[1].imshow(seg.label)
    axes[1].set_title(f'{row["id"]} - Segmentation ({seg.mask.max()} cells)')
    plt.tight_layout()
    plt.savefig(os.path.join(outputdir, f'{row["id"]}_segmentation.png'), dpi=100)
    plt.close()
    
    print(f'细胞数: {seg.mask.max()}')
    print(f'========== 完成: {row["id"]} ==========')

print('\n全部样本处理完成！')
```
