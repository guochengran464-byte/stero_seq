```python
import pandas as pd

samples = [
    ('Y10393D8', 'W202603240063157'),
    ('Y40320PC', 'W202603240063121'),
    ('Y40320ND', 'W202603240063078'),
    ('Y10052H8', 'W202603240063031'),
    ('Y10052E3', 'W202603240062997'),
    ('Y10052F1', 'W202603240062964'),
    ('Y10052H4', 'W202603240062927'),
    ('Y01647G5', 'W202603240062892'),
    ('Y01647G8', 'W202603240062792'),
]

df = pd.DataFrame(samples, columns=['sample_id', 'workflow_id'])
df['gem_path'] = df.apply(
    lambda r: f'/data/input/Files/ResultData/Workflow/{r.workflow_id}/result/{r.sample_id}.raw.gem', axis=1
)

csv_path = '/data/work/liver_hcc_spatial/samples.csv'
df.to_csv(csv_path, index=False)
print(df.to_string(index=False))
print(f'\nSaved to {csv_path}')
```

```python
import numpy as np
import pandas as pd
import cv2
import os

CSV_PATH = '/data/work/liver_hcc_spatial/samples.csv'
OUTDIR   = '/data/work/liver_hcc_spatial/01_gem2png'
os.makedirs(OUTDIR, exist_ok=True)

samples = pd.read_csv(CSV_PATH)

for _, row in samples.iterrows():
    gem_path = row['gem_path']
    out_path = f"{OUTDIR}/{row['sample_id']}.mRNA.png"
    
    print(f'Processing {row["sample_id"]} ...', end=' ')
    data = pd.read_csv(gem_path, sep='\t', comment='#')
    data['x'] -= data['x'].min()
    data['y'] -= data['y'].min()
    data = data.groupby(['x', 'y']).sum().reset_index()
    mtx = np.zeros((data['x'].max()+1, data['y'].max()+1))
    mtx[data['x'], data['y']] = data['MIDCount']
    mtx = np.clip(mtx * 250, 0, 255)
    cv2.imwrite(out_path, mtx)
    print(f'done → {out_path}')

print('\nAll done!')
```
